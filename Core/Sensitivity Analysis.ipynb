```python
from sklearn.preprocessing import normalize
from sklearn.model_selection import train_test_split
from sklearn.metrics import normalized_mutual_info_score, adjusted_mutual_info_score
from data import loader, load_preprocessing
from knn_classifier import knn
import matplotlib.pyplot as plt
import distances
import hott
from kcluster import kclustering
import numpy as np
import kcluster
import importlib
importlib.reload(kcluster)
```

```python
data_path = './data/'
data_name = 'bbcsport-emd_tr_te_splitdict2.pkl'
data_matrix = 'bbcsport-emd_tr_te_split.mat'
data = load_preprocessing(data_path + data_name)
p = 1
C = data['cost_T']
Data = data['X']
Data_lda = data['proportions']
y = data['y']
k = np.unique(y).shape[0]
```

```python
print("Processing wassertein")

n_samples = 50
arrDist = np.zeros((n_samples,26))
arrDistInfo = np.zeros((n_samples,26))
arrPval = np.zeros((n_samples,26))
arrDistMNI = np.zeros(n_samples)
arrDistANI = np.zeros(n_samples)

for i in range(n_samples):
    print("i = ", i)
    dataDist, datalabels, barycenters, dataHist, listPvalue, listInfoDist = kcluster.kclustering(k, Data_lda, C, y, max_iter = 25, reg = 0.01, pVal = False, verbose = False)
    arrDist[i] = dataHist
    arrDistInfo[i] = listInfoDist
    #arrPval[i] = listPvalue
    arrDistMNI[i] = normalized_mutual_info_score(y.astype(int), datalabels.astype(int))
    arrDistANI[i] = adjusted_mutual_info_score(y.astype(int), datalabels.astype(int))

print("Mean info :", np.mean(np.nanmin(arrDistInfo, axis = 1)))
print("Std info :", np.std(np.nanmin(arrDistInfo, axis = 1)))
print("Mean NMI :", np.mean(np.nanmax(arrDistMNI)))
print("Std NMI :", np.std(np.nanmin(arrDistMNI)))
print("Mean AMI :", np.mean(np.nanmax(arrDistANI)))
print("Std Ami :", np.std(np.nanmin(arrDistANI)))
```

```python
topics = [2,5,10,15, 20, 30, 40,50, 70]
lMeanDist = []
lStdDist = []
lMeanMNI = []
lStdMNI = []
lMeanANI = []
lStdANI = []
for el in topics:
    print("Topics", el)
    data_path = './data/'
    data_name = 'bbcsport-emd_tr_te_splitdict' + str(el) + '.pkl'
    data_matrix = 'bbcsport-emd_tr_te_split.mat'
    data = load_preprocessing(data_path + data_name)
    p = 1
    C = data['cost_T']
    Data = data['X']
    Data_lda = data['proportions']
    y = data['y']
    k = np.unique(y).shape[0]

    print("Processing wassertein")

    n_samples = 6
    arrDist = np.zeros((n_samples,26))
    arrDistInfo = np.zeros((n_samples,26))
    arrPval = np.zeros((n_samples,26))
    arrDistMNI = np.zeros(n_samples)
    arrDistANI = np.zeros(n_samples)
    i = 0
    while i < n_samples:
        print("i = ", i)
        dataDist, datalabels, barycenters, dataHist, listPvalue, listInfoDist = kcluster.kclustering(k, Data_lda, C, y, max_iter = 25, reg = 0.01, pVal = False, verbose = False)
        if normalized_mutual_info_score(y.astype(int), datalabels.astype(int)) > 0:
            arrDist[i] = dataHist
            arrDistInfo[i] = listInfoDist
            #arrPval[i] = listPvalue
            arrDistMNI[i] = normalized_mutual_info_score(y.astype(int), datalabels.astype(int))
            arrDistANI[i] = adjusted_mutual_info_score(y.astype(int), datalabels.astype(int))
            i += 1
    
    print("Mean info :", np.mean(np.nanmin(arrDistInfo, axis = 1)))
    lMeanDist.append(np.mean(np.nanmin(arrDistInfo, axis = 1)))
    print("Std info :", np.std(np.nanmin(arrDistInfo, axis = 1)))
    lStdDist.append(np.std(np.nanmin(arrDistInfo, axis = 1)))
    print("Mean NMI :", np.mean(arrDistMNI))
    lMeanMNI.append(np.mean(arrDistMNI))
    print("Std NMI :", np.std(arrDistMNI))
    lStdMNI.append(np.std(arrDistMNI))
    print("Mean AMI :", np.mean(arrDistANI))
    lMeanANI.append(np.mean(arrDistANI))
    print("Std Ami :", np.std(arrDistANI))
    lStdANI.append(np.std(arrDistANI))
```

```python
plt.figure(figsize=(15,10))
plt.plot(topics, np.array(lMeanDist), 'or')
plt.plot(topics, np.array(lMeanDist), '-', color='gray')
plt.fill_between(topics, np.array(lMeanDist) - np.array(lStdDist), np.array(lMeanDist) + np.array(lStdDist), color='gray', alpha=0.2)
plt.xlabel("Number of Topics")
plt.ylabel("Variation of Information")
plt.savefig('bbcsportsensInfo.pdf', dpi=800)
```
